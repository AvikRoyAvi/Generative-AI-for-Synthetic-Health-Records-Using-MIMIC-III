```python
import pandas as pd
import os

SYNTHETIC_PATH = r"data\synthetic"
PROCESSED_PATH = r"data\processed"

# Load all three datasets
train_real     = pd.read_csv(os.path.join(PROCESSED_PATH,  'train_real.csv'))
synthetic_ctgan   = pd.read_csv(os.path.join(SYNTHETIC_PATH, 'synthetic_ctgan.csv'))
synthetic_tabddpm = pd.read_csv(os.path.join(SYNTHETIC_PATH, 'synthetic_tabddpm.csv'))

print("REAL TRAIN:")
print(f"  Shape:          {train_real.shape}")
print(f"  Mortality rate: {train_real['HOSPITAL_EXPIRE_FLAG'].mean()*100:.1f}%")

print("\nCTGAN SYNTHETIC:")
print(f"  Shape:          {synthetic_ctgan.shape}")
print(f"  Mortality rate: {synthetic_ctgan['HOSPITAL_EXPIRE_FLAG'].mean()*100:.1f}%")

print("\nTabDDPM SYNTHETIC:")
print(f"  Shape:          {synthetic_tabddpm.shape}")
print(f"  Mortality rate: {synthetic_tabddpm['HOSPITAL_EXPIRE_FLAG'].mean()*100:.1f}%")

print("\nShapes match:", 
      train_real.shape == synthetic_ctgan.shape == synthetic_tabddpm.shape)
```
